### Assignment 1

The company I am trying to do WebScrapping is Flashbay and the review link is the following: https://ca.trustpilot.com/review/flashbay.ca

#### Write a function that iterate each page and get all of the required infomation for each review

import all the required modules

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
res_url = 'https://ca.trustpilot.com/review/flashbay.ca'

web_reviews = []

while len(web_reviews) < 500:
    html_code = requests.get(res_url).text
    soup = BeautifulSoup(html_code)
    # get all the gtml code from the review container of the website page
    review_container = soup.find_all(name='div', attrs={'class':"styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ"})

    for container in review_container:
        try:
            companyName = 'Flashbay'
            datePublished_element = container.find(name='time', attrs={"data-service-review-date-time-ago":"true"})
            ratingValue_element = container.find(name='div', attrs={'class':"styles_reviewHeader__iU9Px"})
            reviewBody_element = container.find(name='p', attrs={'class':"typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn"})
        
            if not (datePublished_element and ratingValue_element and reviewBody_element):
                continue

            datePublished = datePublished_element.attrs['datetime']
            ratingValue = ratingValue_element.attrs['data-service-review-rating']
            reviewBody = reviewBody_element.text

            web_reviews.append([companyName, datePublished, ratingValue, reviewBody])
    
        except AttributeError:
            continue
        
    next_page = soup.find(name='a', attrs={'class':"link_internal__7XN06 button_button__T34Lr button_m__lq0nA button_appearance-outline__vYcdF button_squared__21GoE link_button___108l pagination-link_next__SDNU4 pagination-link_rel__VElFy"})
    if next_page:
        res_url = 'https://ca.trustpilot.com' + next_page.attrs['href']
    else:
        break


web_reviews
```

#### Write a CSV file that includes at least 500 records from the website

```python
with open('reviews_scrapper.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(["companyName", "datePublished", "ratingValue", "reviewBody"])
    writer.writerows(web_reviews)
```
